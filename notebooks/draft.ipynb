```python
def set_idx(df: pd.DataFrame, idx_col: str) -> pd.DataFrame:
    if idx_col in df.columns:
        df = df.set_index(idx_col)
    return df
```

```python
def drop_unnecesary_id(df: pd.DataFrame) -> pd.DataFrame:
    if 'ID_y' in df.columns:
        df = df.drop('ID_y', axis=1)
    return df
```

```python
def add_ord_edu(df: pd.DataFrame) -> pd.DataFrame:
    df[f'{EDU_COL}_ord'] = df[EDU_COL].str.slice(0, 1).astype(np.int8).values
    return df
```

```python
def fill_sex(df: pd.DataFrame) -> pd.DataFrame:
    most_freq = df[SEX_COL].value_counts().index[0]
    df[SEX_COL] = df[SEX_COL].fillna(most_freq)
    return df  
```

```python
def cast_types(df: pd.DataFrame) -> pd.DataFrame:
    df[CAT_COLS] = df[CAT_COLS].astype('category')

    ohe_int_cols = train[OHE_COLS].select_dtypes('number').columns
    df[ohe_int_cols] = df[ohe_int_cols].astype(np.int8)

    df[REAL_COLS] = df[REAL_COLS].astype(np.float32)
    return df
```

```python
def time_diff(start: datetime, end: datetime):
    if isinstance(start, datetime_time): # convert to datetime
        assert isinstance(end, datetime_time)
        start, end = [datetime.combine(datetime.min, t) for t in [start, end]]
    if start <= end: # e.g., 10:33:26-11:15:49
        return end - start
    else: # end < start e.g., 23:55:00-00:25:00
        end += timedelta(1) # +day
        assert end > start
        return end - start
```

```python
def compute_sleep_time(df: pd.DataFrame) -> pd.DataFrame:
    return time_diff(datetime.strptime(df[SLEEP_TIME_COL], '%H:%M:%S'), datetime.strptime(df[WAKE_UP_TIME_COL], '%H:%M:%S')).seconds//3600
```

```python
def add_sleep_time(df: pd.DataFrame) -> pd.DataFrame:
    df[SLEEP_COL] = df[[SLEEP_TIME_COL, WAKE_UP_TIME_COL]].apply(lambda x: compute_sleep_time(x), axis=1)
    return df
```

```python
def remove_sleep_time_column(df: pd.DataFrame) -> pd.DataFrame:
    print(df.columns)
    if SLEEP_TIME_COL in df.columns:
        df = df.drop(SLEEP_TIME_COL, axis='columns')
    if WAKE_UP_TIME_COL in df.columns:
        df = df.drop(WAKE_UP_TIME_COL, axis='columns')
    return df
```

```python
def replace_nan_values(df: pd.DataFrame) -> pd.DataFrame:
    for col in filter(lambda x: x != 'Частота пасс кур', NAN_COLS):
        df[col] = df[col].fillna(0)
    df['Частота пасс кур'] = df['Частота пасс кур'].fillna('0')
    return df
```

```python
def processed(df: pd.DataFrame) -> pd.DataFrame:
    return df.pipe(set_idx, ID_COL).pipe(drop_unnecesary_id).pipe(fill_sex).pipe(replace_nan_values).pipe(cast_types).pipe(add_ord_edu).pipe(add_sleep_time)
```
